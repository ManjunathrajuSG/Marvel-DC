```python
data1=pd.read_json("Aditi.json")
data1.head()
```

```python
accounting_df = pd.json_normalize(data1['accounting'])
accounting_df['department'] = 'accounting'  # Add a department column
sales_df = pd.json_normalize(data1['sales'])
sales_df['department'] = 'sales'  # Add a department column
 
# Combine the two DataFrames
df = pd.concat([accounting_df, sales_df], ignore_index=True)
 
# Display the final DataFrame
df.head()
```

```python
data2=pd.read_json("shopping.json")
data2.head()
```

```python
groceries_df = pd.json_normalize(data2['shoppingList']['groceries'])
groceries_df['category'] = 'groceries' # Add a department column

personalCare_df=pd.json_normalize(data2['shoppingList']['personalCare'])
personalCare_df['category']='personalCare'

householdItems_df=pd.json_normalize(data2['shoppingList']['householdItems'])
householdItems_df['category']='householdItems'

 
# Combine the two DataFrames
df = pd.concat([groceries_df, personalCare_df,householdItems_df], ignore_index=True)
 
# Display the final DataFrame
df.head()
df.tail()
```
