```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
alignment_counts = data.groupby(['ALIGN', 'source']).size().reset_index(name='count of characters')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the data
plt.figure(figsize=(10,6))
sns.barplot(data=alignment_counts, x='ALIGN', y='count of characters', hue='source', palette='Set2')
plt.title('Distribution of Character Alignment in Marvel and DC')
plt.xlabel('Character Alignment')
plt.ylabel('Count of Characters')
plt.legend(title='source')
plt.show()
```

```python
# Filter valid years and genders
gender_data = data[['source', 'YEAR', 'SEX']].dropna()
gender_count = gender_data.groupby(['source', 'YEAR', 'SEX']).size().reset_index(name='Count')

# Plot
plt.figure(figsize=(14, 8))
sns.lineplot(x='YEAR', y='Count', hue='SEX', style='source', data=gender_count, palette='Set2', markers=True)
plt.title('Gender Ratio in Marvel and DC Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Gender & Source')
plt.show()
```

```python
# Group appearances by year and source
appearance_trend = data.groupby(['source', 'YEAR'])['APPEARANCES'].sum().reset_index()

# Plot
plt.figure(figsize=(14, 8))
sns.lineplot(x='YEAR', y='APPEARANCES', hue='source', data=appearance_trend, palette='coolwarm')
plt.title('Trend of Character Appearances Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Appearances')
plt.legend(title='Source')
plt.show()
```

```python
# Count eye colors for Marvel and DC
# 1. Which eye colors are most common among characters?

eye_color_counts = data.groupby(['source', 'EYE']).size().reset_index(name='Count')
top_eye_colors = eye_color_counts.sort_values(by='Count', ascending=False).groupby('source').head(5)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='EYE', y='Count', hue='source', data=top_eye_colors, palette='rocket')
plt.title('Top 5 Most Common Eye Colors in Marvel and DC')
plt.xlabel('Eye Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Source')
plt.show()
```

```python
# What percentage of Marvel/DC characters are Good, Bad, or Neutral?

# Filter Marvel or DC dataset (example: Marvel)
file1_alignment_counts = file1['ALIGN'].value_counts(normalize=True) * 100

# Plot
plt.figure(figsize=(8, 6))
file1_alignment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#66c2a5', '#fc8d62', '#8da0cb'])
plt.title('Character Alignment Distribution (Marvel)')
plt.ylabel('')  # Remove y-axis label
plt.show()
```

```python
#What is the gender breakdown in Marvel/DC characters?
# Filter Marvel or DC dataset (example: Marvel)
file1_gender_counts = file1['SEX'].value_counts()

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x=file1_gender_counts.index, y=file1_gender_counts.values, palette='warm')
plt.title('Gender Distribution (Marvel)')
plt.xlabel('Gender')
plt.ylabel('Number of Characters')
plt.xticks(rotation=45)
plt.show()
```

```python
import sqlite3
```

```python
conn=sqlite3.connect('data.db')
data.to_sql('merged_data',conn,if_exists='replace',index=False)
```

```python
def execute_query(query):
    cursor=conn.cursor()
    cursor.execute(query)
    result=cursor.fetchall()
    cursor.close()
    return result
```

```python
query1="""
SELECT Actual_Name,Source,APPEARANCES,ALIGN
FROM merged_data
WHERE APPEARANCES IS NOT NULL
ORDER BY Source, APPEARANCES DESC
LIMIT 5;
"""
for row in execute_query(query1):
    print(row)
```

```python
query2="""
select COUNT(*) as characters_with_man_Woman
FROM merged_data
where Actual_Name LIKE '%man%' OR Actual_Name LIKE '%Woman%';
"""
print(execute_query(query2)[0][0])
```

```python
query3="""
select COUNT(*)
from merged_data
where HAIR='Blond Hair' AND ALIGN='Good Characters';
"""
print(execute_query(query3)[0][0])
```

```python
query4="""
SELECT Actual_Name, COUNT(DISTINCT FIRST_APPEARANCE) as appearance_variations
FROM merged_data
WHERE Actual_Name IS NOT NULL AND FIRST_APPEARANCE IS NOT NULL
GROUP BY Actual_Name
HAVING COUNT(DISTINCT FIRST_APPEARANCE) > 1;
"""
for row in execute_query(query4):
    print(row)
```
